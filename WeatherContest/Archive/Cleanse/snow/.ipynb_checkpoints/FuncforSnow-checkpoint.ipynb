```python
import numpy as np
import pandas as pd

def DataFrame(df1):
    #변환 전 데이터 정보
    print('     Before df1     \n')
    df1.info()
    print()
    
    # nan값 변환
    for k in ['Humi{}','dew{}']:
        for i in range(len(df1)):
            try:
                if (not np.isnan(df1[k.format(r'.x')][i])) & (not np.isnan(df1[k.format(r'.y')][i])):
                    continue
                elif np.isnan(df1[k.format(r'.x')][i]):    
                    df1[k.format(r'.x')][i] = df1[k.format(r'.y')][i]
                elif np.isnan(df1[k.format(r'.y')][i]):    
                    df1[k.format(r'.y')][i] = df1[k.format(r'.x')][i]
            except:
                continue

    # snow
    for m in range(len(df1)):
        try:
            if (not np.isnan(df1['snow.x'][m])) & (np.isnan(df1['snow.y'][m])):    
                df1['snow.y'][m] = df1['snow.x'][m]
            
            if (not np.isnan(df1['snow.y'][m])) & (np.isnan(df1['snow.x'][m])):    
                df1['snow.x'][m] = df1['snow.y'][m]
        except:
            continue

                
    # 변환 후 데이터 정보            
    print('     after: df1     \n')
    df1.info()
    print()
    
    df1['Humi'] = (df1['Humi.x']+df1['Humi.y'])/2
    df1['dew'] = (df1['dew.x']+df1['dew.y'])/2
    df1['snow'] = (df1['snow.x']+df1['snow.y'])/2
    
    #데이터 출력
    df1.to_csv(folder + r"confile.csv", encoding="ISO-8859-1")
    
    print("Completed!!")
    
    
```

```python
# 불러올 파일 저장된 경로 (경로의 마지막 폴더 이름 바꿔주기)
folder = r'/Users/lesson6_mac/Downloads/TeamProject/weather3/snow/seoul/'
df1 = pd.DataFrame(pd.read_csv(folder + "confile1.csv",encoding="ISO-8859-1"))

# 함수 실행
DataFrame(df1)
```
