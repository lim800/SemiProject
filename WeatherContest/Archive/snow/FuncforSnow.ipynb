```python
import numpy as np
import pandas as pd

def DataFrame(df1, df2):
    #변환 전 데이터 정보
    print('     Before df1     \n')
    df1.info()
    print('\n     Before df2     \n')
    df2.info()
    print()
    
    # nan값 변환
    for k in ['Humi','dew']:
        for i in range(max(len(df1),len(df2))):
            try:
                if (not np.isnan(df1[k][i])) & (not np.isnan(df2[k][i])):
                    
                    continue
                elif np.isnan(df1[k][i]):    
                    df1[k][i] = df2[k][i]
                elif np.isnan(df2[k][i]):    
                    df2[k][i] = df1[k][i]
            except:
                continue

    # snow
    for m in range(max(len(df1),len(df2))):
        try:
            if (not np.isnan(df1['snow'][m])) & (np.isnan(df2['snow'][m])):    
                df2['snow'][m] = df1['snow'][m]
            
            if (not np.isnan(df2['snow'][m])) & (np.isnan(df1['snow'][m])):    
                df1['snow'][m] = df2['snow'][m]
        except:
            continue

                
    # 변환 후 데이터 정보            
    print('     after: df1     \n')
    df1.info()
    print('\n     after: df2     \n')
    df2.info()
    print()
    
    df3 = (pd.concat([df1['Humi'],df1['dew'],df1['snow']], axis=1)+pd.concat([df2['Humi'],df2['dew'],df2['snow']],axis=1))/2

    
    # 변환 된 두 df 합치기
    df4 = pd.concat([df1, df2, df3], axis=1)
    
    #데이터 출력
    df4.to_csv(folder + r"confile.csv", encoding="ISO-8859-1")
    
    print("Completed!!")
    
    
```
