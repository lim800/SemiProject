```python
import numpy as np
import pandas as pd

def DataFrame(df1):
    #변환 전 데이터 정보
    print('     Before df1     \n')
    df1.info()
    print()
    
    # nan값 변환
    for k in ['temp{}','wind{}','rain{}']:
        for i in range(len(df1)):
            try:
                if (not np.isnan(df1[k.format(r'.x')][i])) & (not np.isnan(df1[k.format(r'.y')][i])):
                    continue
                elif np.isnan(df1[k.format(r'.x')][i]):    
                    df1[k.format(r'.x')][i] = df1[k.format(r'.y')][i]
                elif np.isnan(df1[k.format(r'.y')][i]):    
                    df1[k.format(r'.y')][i] = df1[k.format(r'.x')][i]
            except:
                continue
                
    # 변환 후 데이터 정보            
    print('     after: df1     \n')
    df1.info()
    print()
    
    df1['temp'] = (df1['temp.x']+df1['temp.y'])/2
    df1['wind'] = (df1['wind.x']+df1['wind.y'])/2
    df1['rain'] = (df1['rain.x']+df1['rain.y'])/2

    
    #데이터 출력
    df1.to_csv(folder + r"confile.csv", encoding="ISO-8859-1")
    
    print("Completed!!")
```
